### Szkice wykresÃ³w w plotly

```python
fig = px.histogram(pd.concat((bloki1['price'], bloki2['price'])), x='price', nbins=50)
fig.show()
```

```python
fig = px.histogram(pd.concat((bloki1['price_per_m2'], bloki2['price_per_m2'])), x='price_per_m2', nbins=50)
fig.show()
```

```python
# Build year hist
fig = px.histogram(bloki2, x='build_year', nbins=10)
fig.show()
```

```python
# Pie chart liczby pokojÃ³w
fig = px.pie(pd.concat((bloki1["rooms"], bloki2['rooms'].astype(int))), values='rooms', names='rooms')
fig.show()
```

```python
# Obliczenie jak dÅ‚ugo ogÅ‚oszenie wisi w bazie
from datetime import datetime

bloki1['days_passed'] = (bloki1['lastRefresh'].apply(datetime.fromisoformat) - bloki1['createdTime'].apply(datetime.fromisoformat)).dt.days
bloki2['days_passed'] = (bloki2['date_lastmod'].apply(datetime.fromisoformat) - bloki2['date_created'].apply(datetime.fromisoformat)).dt.days
```

```python
fig = px.histogram(pd.concat((bloki1['days_passed'], bloki2['days_passed'])), x='days_passed', nbins=50)
fig.show()
```
